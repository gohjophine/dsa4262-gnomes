```python
import json
import pandas as pd
import numpy as np
```

# Parse JSON File

```python
df = pd.read_json('../data/data.json', lines=True)
```

```python
df
```

```python
# column of ENST00000000233 after dropping the null values
test = df['ENST00000000233'].dropna()
df2 = pd.DataFrame(test)
df2
```

```python
# return the key and value of the dictionary separately
# input: dictionary
# output: key, value
def split(x):
    key = list(x.keys())[0]
    value = x.get(key)
    return key, value
```

```python
# expand the nested dictionary
transcript_list, position_list, fivemers_list, reads_list = [], [], [], []
for colname in df.columns:
    test = df[colname].dropna()
    for row in test: # each row is a dictionary
        position, dic = split(row)
        fivemers, reads = split(dic)
        
        transcript_list.append(colname)
        position_list.append(position)
        fivemers_list.append(fivemers)
        reads_list.append(reads)
```

```python
df = pd.DataFrame()
df['transcript_id'] = transcript_list
df['transcript_position'] = position_list
df['fivemers'] = fivemers_list
df['reads'] = reads_list
df
```

```python
df = df.explode('reads') # each row is one read
df
```

# Export to csv

```python
df.to_csv('../data/data.csv', index=False)
```
